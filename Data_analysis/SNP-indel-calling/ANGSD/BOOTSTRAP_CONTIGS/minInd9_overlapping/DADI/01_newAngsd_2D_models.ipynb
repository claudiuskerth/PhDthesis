```python
def split_nomig(params, ns, pts):
    """
    params = (nu1,nu2,T)
    ns = (n1,n2)

    Split into two populations of specifed size, no migration.

    nu1: Size of population 1 after split.
    nu2: Size of population 2 after split.
    T: Time in the past of split (in units of 2*Na generations) 
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,T = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi)

    phi = dadi.Integration.two_pops(phi, xx, T, nu1, nu2, m12=0, m21=0)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

```python
def split_nomig(params, ns, pts):
    """
    params = (nu1,nu2,T)
    ns = (n1,n2)

    Split into two populations of specifed size, no migration.

    nu1: Size ratio of population 1 after split (with respect to ancestral population size Na)
    nu2: Size ratio of population 2 after split (with respect to ancestral population size Na)
    T: Time in the past of split (in units of 2*Na generations) 
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,T = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi)

    phi = dadi.Integration.two_pops(phi, xx, T, nu1, nu2, m12=0, m21=0)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

```python
def split_asym_mig(params, ns, pts):
    """
    params = (nu1,nu2,T,m1,m2)
    ns = (n1,n2)

    Split into two populations of specifed size, with potentially asymmetric migration.

    nu1: population size ratio of population 1 after split (with respect to Na)
    nu2: population size ratio of population 2 after split (with respect to Na) 
    T: Time in the past of split (in units of 2*Na generations) 
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,T,m1,m2 = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    
    # split
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) 
    
    # divergence with potentially asymmetric migration
    phi = dadi.Integration.two_pops(phi, xx, T, nu1, nu2, m12=m2, m21=m1)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

## LRT

Are the two migration rates significantly different? I think it is difficult to do an LRT with the current parameterisation. The model with one (symmetrical) migration rate is *not* just nested within the model with two migration rates. One can, however, re-parameterise the asymmetric migration model so that the divergence with (symmetrical) migration model is nested within it by setting $m_2 = r \times m_1$. The divergence with symmetrical migration has an $r$ of 1. Then I can ask: is $r$ significantly different from 1?

```python
def split_asym_mig_m2ratio(params, ns, pts):
    """
    params = (nu1,nu2,T,m1,m2)
    ns = (n1,n2)

    Split into two populations of specifed size, with potentially asymmetric migration.

    nu1: population size ratio of population 1 after split (with respect to Na)
    nu2: population size ratio of population 2 after split (with respect to Na) 
    T: Time in the past of split (in units of 2*Na generations) 
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,T,m1,r = params
    m2 = r * m1
    
    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    
    # split
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) 
    
    # divergence with potentially asymmetric migration
    phi = dadi.Integration.two_pops(phi, xx, T, nu1, nu2, m12=m2, m21=m1)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

# two epochs for $N_e$

```python
def split_asym_mig_2epoch(params, ns, pts):
    """
    params = (nu1_1,nu2_1,T1,nu1_2,nu2_2,T2,m1,m2)
    ns = (n1,n2)

    Split into two populations of specified size, with potentially asymmetric migration.
    The split coincides with a stepwise size change in the daughter populations. Then,
    have a second stepwise size change at some point in time after the split. This is
    enforced to happen at the same time for both populations. Migration is assumed to
    be the same during both epochs.

    nu1_1: pop size ratio of pop 1 after split (with respect to Na)
    nu2_1: pop size ratio of pop 2 after split (with respect to Na)
    T1: Time from split to second size change (in units of 2*Na generations)
    nu1_2: pop size ratio of pop 1 after second size change (with respect to Na)
    nu2_2: pop size ratio of pop 2 after second size change (with respect to Na)
    T2: time in past of second size change (in units of 2*Na generations)
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1_1,nu2_1,T1,nu1_2,nu2_2,T2,m1,m2 = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    
    # split
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) 
    
    # divergence with potentially asymmetric migration for time T1
    phi = dadi.Integration.two_pops(phi, xx, T1, nu1_1, nu2_1, m12=m2, m21=m1)
    
    # divergence with potentially asymmetric migration and different pop size for time T2
    phi = dadi.Integration.two_pops(phi, xx, T2, nu1_2, nu2_2, m12=m2, m21=m1)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

## LRT

This two epoch model can be reduced to the asymmetric migration model (one epoch) by either setting $ery_2$ and $par_2$ equal to $ery_1$ and $par_1$ (i. e. same population sizes in both epochs) or by setting $T_2$ to zero.

I am faced with the problem of having to set parameters equal to other parameters in order to reduce this complex to a more simple model. I am therefore going to re-parameterise this model in a similar way as for the LRT of asymmetric migration vs. symmetric migration models: I am going to redefine $ery_2$ as $ery_r \times ery_1$ and $par_2$ as $par_r \times par_1$.

```python
def split_asym_mig_2epoch_nuRatio(params, ns, pts):
    """
    params = (nu1_1,nu2_1,T1,nu1_2,nu2_2,T2,m1,m2)
    ns = (n1,n2)

    Split into two populations of specified size, with potentially asymmetric migration.
    The split coincides with a stepwise size change in the daughter populations. Then,
    have a second stepwise size change at some point in time after the split. This is
    enforced to happen at the same time for both populations. Migration is assumed to
    be the same during both epochs.

    nu1_1: pop size ratio of pop 1 after split (with respect to Na)
    nu2_1: pop size ratio of pop 2 after split (with respect to Na)
    T1: Time from split to second size change (in units of 2*Na generations)
    nu1_2: pop size ratio of pop 1 after second size change (with respect to Na)
    nu2_2: pop size ratio of pop 2 after second size change (with respect to Na)
    T2: time in past of second size change (in units of 2*Na generations)
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1_1,nu2_1,T1,nu1_r,nu2_r,T2,m1,m2 = params
    nu1_2 = nu1_r * nu1_1
    nu2_2 = nu2_r * nu2_1

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    
    # split
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) 
    
    # divergence with potentially asymmetric migration for time T1
    phi = dadi.Integration.two_pops(phi, xx, T1, nu1_1, nu2_1, m12=m2, m21=m1)
    
    # divergence with potentially asymmetric migration and different pop size for time T2
    phi = dadi.Integration.two_pops(phi, xx, T2, nu1_2, nu2_2, m12=m2, m21=m1)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

# three epochs for $N_e$

I want to give dadi the chance to infer the recent exponential population size decline inferred by `stairway-plot`. I am therefore going to try and fit a 3-epoch model to the data. This model has 11 parameters and it may already be at risk  of overfitting the data, i. e. fitting noise instead of true signal. I therefore must be very careful with the interpretation of the result.

The 3-epoch model is the same as the 2-epoch model for the first two epochs. The third, most recent, epoch does not start with an instantaneous population size change but I am enforcing an exponential pop size change from the population size of the second epoch to the current population size over the time course of the third epoch.

```python
def split_asym_mig_3epoch(params, ns, pts):
    """
    params = (Nery_1,Npar_1,T1,Nery_2,Npar_2,T2,Nery_3,Npar_3,T3,m1,m2)
    ns = (n1,n2)

    Split into two populations of specified size, with potentially asymmetric migration.
    The split coincides with a stepwise size change in the daughter populations. Then,
    have a second stepwise size change at time T2+T3 after the split. Each pop
    size change is enforced to happen at the same time for both populations. Migration 
    is assumed to be the same during all epochs. At time T3 in the past a third pop size 
    change happens. This time it is not instantaneous, but exponentially reaching the current
    pop size over time T3.

    Nery_1: pop size ratio of pop 1 after split (with respect to Na)
    Npar_1: pop size ratio of pop 2 after split (with respect to Na)
    T1: Time from split to second size change (in units of 2*Na generations)
    Nery_2: pop size ratio of pop 1 after second size change (with respect to Na)
    Npar_2: pop size ratio of pop 2 after second size change (with respect to Na)
    T2: time in past of second size change (in units of 2*Na generations)
    Nery_3: pop size ratio of pop 1 after second size change (with respect to Na)
    Npar_3: pop size ratio of pop 2 after second size change (with respect to Na)
    T3: time in past of second size change (in units of 2*Na generations)
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    Nery_1,Npar_1,T1,Nery_2,Npar_2,T2,Nery_3,Npar_3,T3,m1,m2 = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    
    # split
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) 
    
    # divergence with potentially asymmetric migration for time T1
    phi = dadi.Integration.two_pops(phi, xx, T1, Nery_1, Npar_1, m12=m2, m21=m1)
    
    # divergence with potentially asymmetric migration and different pop size for time T2
    phi = dadi.Integration.two_pops(phi, xx, T2, Nery_2, Npar_2, m12=m2, m21=m1)
    
    # exponential growth from Nery_2 to Nery_3 over time T3
    Nery_func = lambda t: Nery_2 * (Nery_3/Nery_2)**(t/T3)
    Npar_func = lambda t: Npar_2 * (Npar_3/Npar_2)**(t/T3)
    
    # divergence with potentially asymmetric migration and 
    # exponentially changing pop size over time T3
    phi = dadi.Integration.two_pops(phi, xx, T3, Nery_func, Npar_func, m12=m2, m21=m1)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

# two epoch with exp. size change

This model is similar to the two epoch model above. It has two epochs with potentially different $N_e$, but instead of two instantaneous population size changes it specifies the second population size change to be exponential and to happen during the whole second (most recent) epoch. This is another test of whether dadi can infer a recent bottleneck in both populations as did `stairway-plot`.

```python
def two_epoch_exp(params, ns, pts):
    """
    params = (Nery_1,Npar_1,T1,Nery_2,Npar_2,T2,m1,m2)
    ns = (n1,n2)

    Split into two populations of specified size, with potentially asymmetric migration.
    The split coincides with a stepwise size change in the daughter populations. Then,
    have a second size change starting at time T2 in the past and exponentially reaching the
    current population size in the present. Each pop size change is enforced to happen at 
    the same time for both populations. Migration is assumed to be the same during all epochs.  

    Nery_1: pop size ratio of pop 1 after split (with respect to Na)
    Npar_1: pop size ratio of pop 2 after split (with respect to Na)
    T1: Time from split to second size change (in units of 2*Na generations)
    Nery_2: pop size ratio of pop 1 at present (with respect to Na)
    Npar_2: pop size ratio of pop 2 at present (with respect to Na)
    T2: time in past of second size change (in units of 2*Na generations)
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    Nery_1,Npar_1,T1,Nery_2,Npar_2,T2,m1,m2 = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    
    # split
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) 
    
    # divergence with potentially asymmetric migration for time T1
    phi = dadi.Integration.two_pops(phi, xx, T1, Nery_1, Npar_1, m12=m2, m21=m1)
    
    # exponential growth/decline from Nery_1 to Nery_2 over time T2
    Nery_func = lambda t: Nery_1 * (Nery_2/Nery_1)**(t/T2)
    Npar_func = lambda t: Npar_1 * (Npar_2/Npar_1)**(t/T2)
    
    # divergence with potentially asymmetric migration and 
    # exponentially changing pop size over time T2
    phi = dadi.Integration.two_pops(phi, xx, T2, Nery_func, Npar_func, m12=m2, m21=m1)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

### define model function for `moments`

```python
def two_epoch_exp_moments(params, ns):
    """
    params = (Nery_1,Npar_1,T1,Nery_2,Npar_2,T2,m1,m2)
    ns = (n1,n2)

    Split into two populations of specified size, with potentially asymmetric migration.
    The split coincides with a stepwise size change in the daughter populations. Then,
    have a second size change starting at time T2 in the past and exponentially reaching the
    current population size in the present. Each pop size change is enforced to happen at 
    the same time for both populations. Migration is assumed to be the same during all epochs.  

    Nery_1: pop size ratio of pop 1 after split (with respect to Na)
    Npar_1: pop size ratio of pop 2 after split (with respect to Na)
    T1: Time from split to second size change (in units of 2*Na generations)
    Nery_2: pop size ratio of pop 1 at present (with respect to Na)
    Npar_2: pop size ratio of pop 2 at present (with respect to Na)
    T2: time in past of second size change (in units of 2*Na generations)
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    Nery_1,Npar_1,T1,Nery_2,Npar_2,T2,m1,m2 = params

    # fs for equilibrium ancestral population
    sts = moments.LinearSystem_1D.steady_state_1D(ns[0] + ns[1])
    fs = moments.Spectrum(sts)
    
    # split
    fs = moments.Manips.split_1D_to_2D(fs, ns[0], ns[1])  
    
    # divergence with potentially asymmetric migration
    fs.integrate([Nery_1, Npar_1], T1, m=np.array([[0, m2], [m1, 0]]))
    
    # exponential growth/decline from Nery_1 to Nery_2 over time T2
    Nery_func = lambda t: Nery_1 * (Nery_2/Nery_1)**(t/T2)
    Npar_func = lambda t: Npar_1 * (Npar_2/Npar_1)**(t/T2)
    nu_func = lambda t: [Nery_func(t), Npar_func(t)]
    
    # divergence with potentially asymmetric migration and 
    # exponentially changing pop size over time T2
    fs.integrate(nu_func, T2, m=np.array([[0, m2], [m1, 0]]))

    return fs
```

# Secondary contact model

In comparison to the _divergence with migration_ model above, can I improve the fit of the data to the model by introducing an initial period without gene flow after the split followed by a period with gene flow?

```python
def secondary_contact(params, ns, pts):
    """
    params = (nu1,nu2,Ti,Tc,m)
    ns = (n1,n2)

    Split into two populations of specified size, 
    then diverge in isolation,
    finally come into secondary contact with **symmetrical** migration.

    nu1: Size of population 1 after split.
    nu2: Size of population 2 after split.
    Ti: Time of divergence in isolation (in units of 2*Na generations)
    Tc: Time in the past when secondary contact began (and continues till present)
    m: symmetrical migration rate (in units of 2*Na per generation)
    
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,Ti,Tc,m = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) # split

    # divergence in isolation
    phi = dadi.Integration.two_pops(phi, xx, Ti, nu1, nu2, m12=0, m21=0) 
    
    # secondary contact
    phi = dadi.Integration.two_pops(phi, xx, Tc, nu1, nu2, m12=m, m21=m)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

# Ancient migration

This model is very similar to the previous _secondary contact_ model. It simply defines the first epoch as the one where gene flow is allowed, followed by a second epoch with complete isolation. It could therefore also be called _primary contact_ model.

```python
def primary_contact(params, ns, pts):
    """
    params = (nu1,nu2,Ti,Tc,m)
    ns = (n1,n2)

    Split into two populations of specified size, 
    then diverge with **symmetrical** gene flow,
    finally diverge in complete isolation.

    nu1: Size of population 1 after split.
    nu2: Size of population 2 after split.
    Tc: Time of primary contact after split (and continues till present)
    Ti: Time of divergence in isolation (in units of 2*Na generations)
    m: symmetrical migration rate (in units of 2*Na per generation)
    
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,Tc,Ti,m = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) # split
        
    # primary contact
    phi = dadi.Integration.two_pops(phi, xx, Tc, nu1, nu2, m12=m, m21=m)

    # divergence in isolation
    phi = dadi.Integration.two_pops(phi, xx, Ti, nu1, nu2, m12=0, m21=0) 

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

# migration + isolation + secondary contact

This is an extension of the previous _ancient migration_ model. It simply adds a third, most recent, epoch where gene flow recommences. In order to make this model comparable to the _divergence with migration_, _secondary contact_ and _ancient migration_ model, I am specifying symmetrical gene flow, although the _asymmetric migration_ model has already shown that asymmetric gene flow significantly improves the fit. As with these other models, population size is assumed to be constant during these three epochs.

```python
def split_mig_iso_mig(params, ns, pts):
    """
    params = (nu1,nu2,Tc,Ti,Tsc,m)
    ns = (n1,n2)

    Split into two populations of specifed size, with **symmetric** migration
    for a time Tc followed by a period of complete isolation Ti which is followed by a restart of
    migration until present. Migration rates in the two epochs are assumed be to equal (and constant).

    nu1: Size of population 1 after split.
    nu2: Size of population 2 after split.
    Tc: Time of gene flow after split (in units of 2*Na generations) 
    m: Migration rate (in units of 2*Na alleles?/ind.? per generation)
    Ti: Time of isolation after cessation of gene flow
    Tsc: Time of gene flow after period of isolation until present
    
    The split lies Tc+Ti+Tsc * 2Na generations in the past.
    
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,Tc,Ti,Tsc,m = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    
    # split
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) 
    
    # divergence with symmetric migration
    phi = dadi.Integration.two_pops(phi, xx, Tc, nu1, nu2, m12=m, m21=m)
    
    # divergence without gene flow
    phi = dadi.Integration.two_pops(phi, xx, Ti, nu1, nu2, m12=0, m21=0)
    
    # divergence with symmetric migration
    phi = dadi.Integration.two_pops(phi, xx, Tsc, nu1, nu2, m12=m, m21=m)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

# ancient asymmetric migration

I am now going to add asymmetric migration to the previous _ancient migration_ model.

```python
def ancient_migration(params, ns, pts):
    """
    params = (nu1,nu2,Ti,Tc,m1,m2)
    ns = (n1,n2)

    Split into two populations of specified size, 
    then diverge with **asymmetrical** gene flow,
    finally diverge in complete isolation.

    nu1: Size of population 1 after split.
    nu2: Size of population 2 after split.
    Tc: Time of primary contact after split (and continues till present)
    Ti: Time of divergence in isolation (in units of 2*Na generations)
    m1: migration rate from ERY into PAR (in units of 2*Na per generation)
    m2: migration rate from PAR into ERY (in units of 2*Na per generation)
    
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,Tc,m1,m2,Ti = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) # split
        
    # primary contact with asymmetric migration
    phi = dadi.Integration.two_pops(phi, xx, Tc, nu1, nu2, m12=m2, m21=m1)

    # divergence in isolation
    phi = dadi.Integration.two_pops(phi, xx, Ti, nu1, nu2, m12=0, m21=0) 

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

```python
def ancient_migration_moments(params, ns): # note, I have removed the pts argument from the argument list
    """
    params = (nu1,nu2,Tc,m1,m2,Ti)
    ns = (n1,n2)

    Split into two populations of specifed size, with potentially asymmetric migration
    for a time Tc followed by complete isolation until present.

    nu1: population size ratio of population 1 after split (with respect to Na)
    nu2: population size ratio of population 2 after split (with respect to Na)
    Tc: Time of gene flow after split (in units of 2*Na generations) 
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    Ti: Time of isolation after cessation of gene flow
    The split lies Tc+Ti * 2Na generations in the past.
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,Tc,m1,m2,Ti = params
    
    # fs for equilibrium ancestral population
    sts = moments.LinearSystem_1D.steady_state_1D(ns[0] + ns[1])
    fs = moments.Spectrum(sts)
    
    # split
    fs = moments.Manips.split_1D_to_2D(fs, ns[0], ns[1]) 
    
    # divergence with potentially asymmetric migration
    fs.integrate([nu1, nu2], Tc, m=np.array([[0, m2], [m1, 0]]))
    
    # divergence without gene flow
    fs.integrate([nu1, nu2], Ti, m=np.array([[0,0], [0,0]]))

    return fs
```

# two epoch migration

I am going to try and allow different rates of gene flow in the two epochs. Population sizes are assumed to be the same during both epochs. This model is therefore a slight extension to the _ancient migration_ model, which enforced 0 gene flow during the second epoch.

```python
def two_epoch_migration(params, ns, pts):
    """
    params = (nu1,nu2,T1,m1,m2,T2,m1_2,m2_2)
    ns = (n1,n2)

    Split into two populations of specifed size, with potentially asymmetric migration
    for a time T1 followed by a second epoch T2 of potentially changed rates of gene flow
    between the two populations.

    nu1: population size ratio of population 1 after split (with respect to Na)
    nu2: population size ratio of population 2 after split (with respect to Na)
    T1: Time of first epoch after split (in units of 2*Na generations) 
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    T2: Time of second epoch
    m1_2: Migration rate from ery into par in 2nd epoch
    m2_2: Migration rate from par into ery in 2nd epoch
    The split lies T1+T2 * 2Na generations in the past.
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,T1,m1,m2,T2,m1_2,m2_2 = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    
    # split
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) 
    
    # divergence with potentially asymmetric migration
    phi = dadi.Integration.two_pops(phi, xx, T1, nu1, nu2, m12=m2, m21=m1)
    
    # divergence with potentially different rate of gene flow
    phi = dadi.Integration.two_pops(phi, xx, T2, nu1, nu2, m12=m2_2, m21=m1_2)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

## for _ancient asymmetric migration_ model

```python
def ancient_migration(params, ns, pts):
    """
    params = (nu1,nu2,Ti,Tc,m1,m2)
    ns = (n1,n2)

    Split into two populations of specified size, 
    then diverge with **asymmetrical** gene flow,
    finally diverge in complete isolation.

    nu1: Size of population 1 after split.
    nu2: Size of population 2 after split.
    Tc: Time of primary contact after split (and continues till present)
    Ti: Time of divergence in isolation (in units of 2*Na generations)
    m1: migration rate from ERY into PAR (in units of 2*Na per generation)
    m2: migration rate from PAR into ERY (in units of 2*Na per generation)
    
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,Tc,m1,m2,Ti = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) # split
        
    # primary contact with asymmetric migration
    phi = dadi.Integration.two_pops(phi, xx, Tc, nu1, nu2, m12=m2, m21=m1)

    # divergence in isolation
    phi = dadi.Integration.two_pops(phi, xx, Ti, nu1, nu2, m12=0, m21=0) 

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

The sign change makes $\hat{V}(xy)$ an unbiased estimator of $V(xy)$.

$x$ and $y$ are assumed to be bivariate  normally distributed. $E(x)$ and $E(y)$ are the maximum likelihood estimates returned by dadi. $C()$ is the covariance.

## for two epoch with exp. size change model

```python
def two_epoch_exp(params, ns, pts):
    """
    params = (Nery_1,Npar_1,T1,Nery_2,Npar_2,T2,m1,m2)
    ns = (n1,n2)

    Split into two populations of specified size, with potentially asymmetric migration.
    The split coincides with a stepwise size change in the daughter populations. Then,
    have a second size change starting at time T2 in the past and exponentially reaching the
    current population size in the present. Each pop size change is enforced to happen at 
    the same time for both populations. Migration is assumed to be the same during all epochs.  

    Nery_1: pop size ratio of pop 1 after split (with respect to Na)
    Npar_1: pop size ratio of pop 2 after split (with respect to Na)
    T1: Time from split to second size change (in units of 2*Na generations)
    Nery_2: pop size ratio of pop 1 at present (with respect to Na)
    Npar_2: pop size ratio of pop 2 at present (with respect to Na)
    T2: time in past of second size change (in units of 2*Na generations)
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    Nery_1,Npar_1,T1,Nery_2,Npar_2,T2,m1,m2 = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    
    # split
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) 
    
    # divergence with potentially asymmetric migration for time T1
    phi = dadi.Integration.two_pops(phi, xx, T1, Nery_1, Npar_1, m12=m2, m21=m1)
    
    # exponential growth/decline from Nery_1 to Nery_2 over time T2
    Nery_func = lambda t: Nery_1 * (Nery_2/Nery_1)**(t/T2)
    Npar_func = lambda t: Npar_1 * (Npar_2/Npar_1)**(t/T2)
    
    # divergence with potentially asymmetric migration and 
    # exponentially changing pop size over time T2
    phi = dadi.Integration.two_pops(phi, xx, T2, Nery_func, Npar_func, m12=m2, m21=m1)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

### get optimal parameter values

```python
def split_nomig(params, ns, pts):
    """
    params = (nu1,nu2,T)
    ns = (n1,n2)

    Split into two populations of specifed size, no migration.

    nu1: Size ratio of population 1 after split (with respect to ancestral population size Na)
    nu2: Size ratio of population 2 after split (with respect to ancestral population size Na)
    T: Time in the past of split (in units of 2*Na generations) 
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,T = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi)

    phi = dadi.Integration.two_pops(phi, xx, T, nu1, nu2, m12=0, m21=0)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

```python
ns = sfs2d.sample_sizes

pts_l = [40, 50, 60]

func_ex = dadi.Numerics.make_extrap_log_func(split_nomig)
```

```python
def split_asym_mig(params, ns, pts):
    """
    params = (nu1,nu2,T,m1,m2)
    ns = (n1,n2)

    Split into two populations of specifed size, with potentially asymmetric migration.

    nu1: population size ratio of population 1 after split (with respect to Na)
    nu2: population size ratio of population 2 after split (with respect to Na) 
    T: Time in the past of split (in units of 2*Na generations) 
    m1: Migration rate from ery into par (in units of 2*Na ind per generation)
    m2: Migration rate from par into ery (in units of 2*Na ind per generation)
    n1,n2: Sample sizes of resulting Spectrum
    pts: Number of grid points to use in integration.
    """
    nu1,nu2,T,m1,m2 = params

    xx = dadi.Numerics.default_grid(pts)

    phi = dadi.PhiManip.phi_1D(xx)
    
    # split
    phi = dadi.PhiManip.phi_1D_to_2D(xx, phi) 
    
    # divergence with potentially asymmetric migration
    phi = dadi.Integration.two_pops(phi, xx, T, nu1, nu2, m12=m2, m21=m1)

    fs = dadi.Spectrum.from_phi(phi, ns, (xx,xx))
    return fs
```

```python
func_ex = dadi.Numerics.make_extrap_log_func(split_asym_mig)
```
