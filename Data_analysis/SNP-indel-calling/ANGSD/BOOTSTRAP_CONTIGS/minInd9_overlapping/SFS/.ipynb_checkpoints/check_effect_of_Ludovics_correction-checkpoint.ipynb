```python
fs_ery = dadi.Spectrum.from_file("original/ERY/ERY.unfolded.sfs.dadi")
fs_ery = fs_ery.fold()
```

```python
# make the extrapolating version of the standard neutral model function
func_ex = dadi.Numerics.make_extrap_log_func(dadi.Demographics1D.snm)

# setting the smallest grid size slightly larger than the largest population sample size
pts_l = [40, 50, 60]

ns = fs_ery.sample_sizes

# calculate unfolded AFS under standard neutral model (up to a scaling factor theta)
snm = func_ex(0, ns, pts_l)
snm = snm.fold()

# scaled snm spectrum for ERY
snm_ery = dadi.Inference.optimally_scaled_sfs(snm, fs_ery)
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.figure(figsize=(6,5))

plt.plot(fs_ery, 'ro-', label='ery')
plt.plot(snm_ery, 'bo-', label='snm')
plt.xlabel('minor allele frequency')
plt.ylabel('SNP count')
plt.grid()
plt.legend()
```

```python
# turn on floating point division by default, old behaviour via '//'
from __future__ import division
```

```python
# define function that modifies a SFS with a given p
def modify(p, sfs):
    """
    modifies SFS with given p
    subtracts proportion p from even count classes and adds to corresponding 
    haploid count class, e. g. from 2 onto 1 or from 16 onto 8
    returns modified sfs
    """
    fs = sfs.copy() # it is necessary to make sure that a copy of the observed SFS is
    # made here for each function evaluation
    
    # modify observed SFS with given p:
    for i in range(len(fs)):
        if not fs.mask[i]:
            if not i%2:
                deduct = fs[i] * p
                fs[i] -= deduct
                j = i//2
                fs[j] += deduct
    return fs
```

```python
# define cost function used for optimisation
def f(p, snm, sfs):
    """
    p: proportion of all SNP's on the X chromosome [float, 0<p<1]
    snm: standard neutral model spectrum (optimally scaled)
    sfs: observed SFS 
    """
    
    # modify sfs
    fs = modify(p, sfs)
            
    # return sum of squared Poisson residuals of modified SFS with snm spectrum:
    # return np.sum( (fs - snm)**2/snm )  # note, due to the mask it is necessary to use the numpy sum function
    # this down-weights residuals in lower frequency classes

    # return sum of squared deviations of modified SFS with snm spectrum:
    return np.sum( (fs - snm)**2 )  # note, due to the mask it is necessary to use the numpy sum function 
```

```python
from scipy.optimize import minimize_scalar
```

```python
# test optimisation
# the function to optimise (cost function) 'f' gets the optimally scaled standard neutral SFS
# and the standard neutral SFS to modify
# I want to see an optimal value for p of zero
res_test = minimize_scalar(f, method = 'bounded', bounds = (-1.0, 1.0), args = (snm_ery, snm_ery))
```

```python
res_test
```

This makes sense: the optimal value is practically zero.

```python
res_ery = minimize_scalar(f, method = 'bounded', bounds = (0.0, 1.0), args = (snm_ery, fs_ery))
```

```python
res_ery
```

The optimal degree of modification is small, only about 7%. This seems credible since the observed 1D SFS of ERY is already quite close to the standard neutral spectrum.

```python
plt.figure(figsize=(6,5))

plt.plot(fs_ery, 'ro-', label='original')
plt.plot(modify(res_ery.x, fs_ery), 'bo-', label='corrected, p=' + str(round(res_ery.x,2)))
plt.xlabel('minor allele frequency')
plt.ylabel('SNP count')
plt.grid()
plt.legend()
plt.title('1D spectrum of ERY')
```

# PAR

```python
fs_par = dadi.Spectrum.from_file("original/PAR/PAR.unfolded.sfs.dadi")
fs_par = fs_par.fold()
```

```python
# make the extrapolating version of the standard neutral model function
func_ex = dadi.Numerics.make_extrap_log_func(dadi.Demographics1D.snm)

# setting the smallest grid size slightly larger than the largest population sample size
pts_l = [40, 50, 60]

ns = fs_par.sample_sizes

# calculate unfolded AFS under standard neutral model (up to a scaling factor theta)
snm = func_ex(0, ns, pts_l)
snm = snm.fold()

# scaled snm spectrum for ERY
snm_par = dadi.Inference.optimally_scaled_sfs(snm, fs_par)
```

```python
plt.figure(figsize=(6,5))

plt.plot(fs_par, 'g^-', label='original')
plt.plot(snm_par, 'bo-', label='snm')
plt.xlabel('minor allele frequency')
plt.ylabel('SNP count')
plt.grid()
plt.legend()
plt.title('1D spectrum of PAR')
```

```python
res_par = minimize_scalar(f, method = 'bounded', bounds = (0.0, 1.0), args = (snm_par, fs_par))
```

```python
res_par
```

The optimal degree of correction is much higher than for ERY, about 15%.

```python
plt.figure(figsize=(6,5))

plt.plot(fs_par, 'g^-', label='original')
plt.plot(modify(res_par.x, fs_par), 'bo-', label='corrected, p=' + str(round(res_par.x,2)))
plt.plot(snm_par, marker=".", linestyle="-", linewidth="1", color='gray', label='snm')
plt.xlabel('minor allele frequency')
plt.ylabel('SNP count')
plt.grid()
plt.legend()
plt.title('1D spectrum of PAR')
```
