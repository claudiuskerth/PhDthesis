Now let's plot the above figures together in one tableau.

```python
import matplotlib as mpl
```

```python
plt.figure(figsize = (16, 14))
MAF_by_pval = ["MAF_by_pval_ery", "MAF_by_pval_par", "MAF_by_pval_negFis_ery", "MAF_by_pval_negFis_par"]
head = ["positive F, ERY", "positive F, PAR", "negative F, ERY", "negative F, PAR"]
for i,Mbp in enumerate(MAF_by_pval):
    pval, MAF, numSNP = [], [], []
    with open(Mbp) as f:
        f.readline() # read the first line, but discard (header)
        for line in f:
            one, two, three = line.strip().split("\t")
            pval.append(float(one))
            MAF.append(float(two))
            numSNP.append(int(three))
    numSNP = np.array(numSNP)
    plt.subplot(2,2,i+1)
    ax = plt.gca()
    ax.semilogx()
    plt.scatter(pval, MAF, s=80, c=np.log10(numSNP), cmap='coolwarm')
    cb = plt.colorbar() # shrink=0.9, ticks=[1, 2, 3, 4, 5]
#    cb.ax.set_yticklabels(['10', '100', '1000', '10000']) 
    if(i%2):
        cb.set_label(r"$log_{10}$ number of SNP's")
    plt.grid()
    if(i>1):
        plt.xlabel('p-value')
    if(not i%2):
        plt.ylabel('average MAF')
    plt.title(head[i])
    #ax.text(20, 0.445, r"$log_{10}$" + " #SNP")

plt.savefig("MAF_by_pval_all.png")
```
